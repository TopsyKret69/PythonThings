```python
import numpy as np
from numpy import linalg as linalg
import matplotlib.pyplot as plot

dW=np.random.randn(1,100)
plot.plot(dW[0,:])
plot.xlabel("time")
plot.ylabel("White Noise")
plot.show()
W=np.cumsum(dW)/10
plot.plot(W)
plot.xlabel("step")
plot.ylabel("Brownian Motion")
plot.show()
```
